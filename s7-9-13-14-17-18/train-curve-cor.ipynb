```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
file_path = 'trainLogs/cifar/resnet/cor/cor-train-1.0.log'
with open(file_path, 'r') as f:
    lines_cor_1_0 = f.readlines()
lines_cor_1_0 = lines_cor_1_0[9:509]

file_path = 'trainLogs/cifar/resnet/cor/cor-train-2.0.log'
with open(file_path, 'r') as f:
    lines_cor_2_0 = f.readlines()
lines_cor_2_0 = lines_cor_2_0[9:509]

file_path = 'trainLogs/cifar/resnet/cor/cor-train-4.0.log'
with open(file_path, 'r') as f:
    lines_cor_4_0 = f.readlines()
lines_cor_4_0 = lines_cor_4_0[9:509]

file_path = 'trainLogs/cifar/resnet/cor/cor-train-8.0.log'
with open(file_path, 'r') as f:
    lines_cor_8_0 = f.readlines()
lines_cor_8_0 = lines_cor_8_0[9:509]
```

```python
train_loss_1_0 = []
test_loss_1_0 = []
test_accuracy_1_0 = []
r_1_0 = []
for i,line in enumerate(lines_cor_1_0):
    if i % 5 == 1:
        train_loss_1_0.append(line.split()[2])
    if i % 5 == 2:
        r_1_0.append(line.split()[2])
    if i % 5 == 3:
        test_loss_1_0.append(line.split()[2])
    if i % 5 == 4:
        test_accuracy_1_0.append(line.split()[2])
```

```python
train_loss_2_0 = []
test_loss_2_0 = []
test_accuracy_2_0 = []
r_2_0 = []
for i,line in enumerate(lines_cor_2_0):
    if i % 5 == 1:
        train_loss_2_0.append(line.split()[2])
    if i % 5 == 2:
        r_2_0.append(line.split()[2])
    if i % 5 == 3:
        test_loss_2_0.append(line.split()[2])
    if i % 5 == 4:
        test_accuracy_2_0.append(line.split()[2])
```

```python
train_loss_4_0 = []
test_loss_4_0 = []
test_accuracy_4_0 = []
r_4_0 = []
for i,line in enumerate(lines_cor_4_0):
    if i % 5 == 1:
        train_loss_4_0.append(line.split()[2])
    if i % 5 == 2:
        r_4_0.append(line.split()[2])
    if i % 5 == 3:
        test_loss_4_0.append(line.split()[2])
    if i % 5 == 4:
        test_accuracy_4_0.append(line.split()[2])
```

```python
train_loss_8_0 = []
test_loss_8_0 = []
test_accuracy_8_0 = []
r_8_0 = []
for i,line in enumerate(lines_cor_8_0):
    if i % 5 == 1:
        train_loss_8_0.append(line.split()[2])
    if i % 5 == 2:
        r_8_0.append(line.split()[2])
    if i % 5 == 3:
        test_loss_8_0.append(line.split()[2])
    if i % 5 == 4:
        test_accuracy_8_0.append(line.split()[2])
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname="./simhei.ttf")
import pylab as p
```

```python
cor = 1.0
i = 0
shape = ['-k', '--k', '-.k', ':k']
for loss in [test_loss_1_0,test_loss_2_0,test_loss_4_0,test_loss_8_0]:
    plt.plot([x for x in range(100)],[float(x) for x in loss], shape[i] ,label='λ=%s'%str(cor))
    i = i + 1
    cor = cor * 2

plt.xlabel('训练轮数',FontProperties=my_font,fontsize=16)
plt.ylabel('测试集损失值',FontProperties=my_font,fontsize=16)
plt.legend(prop=my_font)
plt.show()
```

```python
cor = 1.0
i = 0
shape = ['-k', '--k', '-.k', ':k']
for loss in [r_1_0,r_2_0,r_4_0,r_8_0]:
    plt.plot([x for x in range(100)],[float(x) for x in loss], shape[i] ,label='λ=%s'%str(cor))
    i = i + 1
    cor = cor * 2

plt.xlabel('训练轮数',FontProperties=my_font,fontsize=16)
plt.ylabel('相关值',FontProperties=my_font,fontsize=16)
plt.legend(prop=my_font)
plt.show()
```

```python
cor = 1.0
i = 0
shape = ['-k', '--k', '-.k', ':k']
for loss in [train_loss_1_0,train_loss_2_0,train_loss_4_0,train_loss_8_0]:
    plt.plot([x for x in range(100)],[float(x) for x in loss], shape[i] ,label='λ=%s'%str(cor))
    i = i + 1
    cor = cor * 2

plt.xlabel('训练轮数',FontProperties=my_font,fontsize=16)
plt.ylabel('训练集损失',FontProperties=my_font,fontsize=16)
plt.legend(prop=my_font)
plt.show()
```

```python
cor = 1.0
i = 0
shape = ['-k', '--k', '-.k', ':k']
for loss in [test_accuracy_1_0,test_accuracy_2_0,test_accuracy_4_0,test_accuracy_8_0]:
    plt.plot([x for x in range(100)],[float(x) for x in loss], shape[i] ,label='λ=%s'%str(cor))
    i = i + 1
    cor = cor * 2

plt.xlabel('训练轮数',FontProperties=my_font,fontsize=16)
plt.ylabel('测试集准确率',FontProperties=my_font,fontsize=16)
plt.legend(prop=my_font)
plt.show()
```
