```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
file_path = 'trainLogs/cifar/resnet/cap/cap-train-0.1.log'
with open(file_path, 'r') as f:
    lines_cap = f.readlines()
lines_cap = lines_cap[9:609]
```

```python
train_loss = []
mal_loss = []
test_loss = []
test_accuracy = []
mal_acc = []
for i,line in enumerate(lines_cap):
    if i % 6 == 1:
        train_loss.append(line.split()[2])
    elif i % 6 == 2:
        mal_loss.append(line.split()[2])
    elif i % 6 == 3:
        mal_acc.append(line.split()[2])
    elif i % 6 == 4:
        test_loss.append(line.split()[2])
    elif i % 6 == 5:
        test_accuracy.append(line.split()[2])
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname="./simhei.ttf")
import pylab as p
```

```python
cor = 1.0
shape = ['-k', '--k', '-.k', ':k']
loss_name = ['训练集', '合成数据集','测试集']
for i,loss in enumerate([train_loss,mal_loss,test_loss]):
    plt.plot([x for x in range(100)],[float(x) for x in loss], shape[i] ,label=loss_name[i])

plt.xlabel('训练轮数',FontProperties=my_font)
plt.ylabel('损失值',FontProperties=my_font)
plt.legend(prop=my_font)
plt.show()
```

```python
cor = 1.0
shape = ['-k', '--k', '-.k', ':k']
loss_name = ['测试集', '合成数据集']
for i,loss in enumerate([test_accuracy,mal_acc]):
    plt.plot([x for x in range(100)],[float(x) for x in loss], shape[i] ,label=loss_name[i])

plt.xlabel('训练轮数',FontProperties=my_font,fontsize=16)
plt.ylabel('准确率',FontProperties=my_font, fontsize=16)
plt.legend(prop=my_font)
plt.show()
```
