```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname="./simhei.ttf")
import pylab as p
import os
```

```python
file_path = 'trainLogs/cifar/resnet/sgn/sgn-train-8.0.log'
with open(file_path, 'r') as f:
    lines_cor_8_0 = f.readlines()
lines_cor_8_0 = lines_cor_8_0[10:510]

file_path = 'trainLogs/cifar/resnet/sgn/sgn-train-16.0.log'
with open(file_path, 'r') as f:
    lines_cor_16_0 = f.readlines()
lines_cor_16_0 = lines_cor_16_0[10:510]

file_path = 'trainLogs/cifar/resnet/sgn/sgn-train-32.0.log'
with open(file_path, 'r') as f:
    lines_cor_32_0 = f.readlines()
lines_cor_32_0 = lines_cor_32_0[10:510]

file_path = 'trainLogs/cifar/resnet/sgn/sgn-train-64.0.log'
with open(file_path, 'r') as f:
    lines_cor_64_0 = f.readlines()
lines_cor_64_0 = lines_cor_64_0[10:510]
```

```python
train_loss_8_0 = []
test_loss_8_0 = []
test_accuracy_8_0 = []
r_8_0 = []
for i,line in enumerate(lines_cor_8_0):
    if i % 5 == 1:
        train_loss_8_0.append(line.split()[2])
    if i % 5 == 2:
        r_8_0.append(line.split()[2])
    if i % 5 == 3:
        test_loss_8_0.append(line.split()[2])
    if i % 5 == 4:
        test_accuracy_8_0.append(line.split()[2])
```

```python
train_loss_16_0 = []
test_loss_16_0 = []
test_accuracy_16_0 = []
r_16_0 = []
for i,line in enumerate(lines_cor_16_0):
    if i % 5 == 1:
        train_loss_16_0.append(line.split()[2])
    if i % 5 == 2:
        r_16_0.append(line.split()[2])
    if i % 5 == 3:
        test_loss_16_0.append(line.split()[2])
    if i % 5 == 4:
        test_accuracy_16_0.append(line.split()[2])
```

```python
train_loss_32_0 = []
test_loss_32_0 = []
test_accuracy_32_0 = []
r_32_0 = []
for i,line in enumerate(lines_cor_32_0):
    if i % 5 == 1:
        train_loss_32_0.append(line.split()[2])
    if i % 5 == 2:
        r_32_0.append(line.split()[2])
    if i % 5 == 3:
        test_loss_32_0.append(line.split()[2])
    if i % 5 == 4:
        test_accuracy_32_0.append(line.split()[2])
```

```python
train_loss_64_0 = []
test_loss_64_0 = []
test_accuracy_64_0 = []
r_64_0 = []
for i,line in enumerate(lines_cor_64_0):
    if i % 5 == 1:
        train_loss_64_0.append(line.split()[2])
    if i % 5 == 2:
        r_64_0.append(line.split()[2])
    if i % 5 == 3:
        test_loss_64_0.append(line.split()[2])
    if i % 5 == 4:
        test_accuracy_64_0.append(line.split()[2])
```

```python
cor = 8.0
i = 0
shape = ['-k', '--k', '-.k', ':k']
for loss in [test_loss_8_0,test_loss_16_0,test_loss_32_0,test_loss_64_0]:
    plt.plot([x for x in range(100)],[float(x) for x in loss], shape[i] ,label='λ=%s'%str(cor))
    i = i + 1
    cor = cor * 2

plt.xlabel('训练轮次',FontProperties=my_font,fontsize=16)
plt.ylabel('测试集损失值',FontProperties=my_font,fontsize=16)
plt.legend(prop=my_font)
plt.show()
```

```python
cor = 8.0
i = 0
shape = ['-k', '--k', '-.k', ':k']
for loss in [r_8_0,r_16_0,r_32_0,r_64_0]:
    plt.plot([x for x in range(100)],[float(x) for x in loss], shape[i] ,label='λ=%s'%str(cor))
    i = i + 1
    cor = cor * 2

plt.xlabel('训练轮次',FontProperties=my_font,fontsize=16)
plt.ylabel('符号相关值',FontProperties=my_font,fontsize=16)
plt.legend(prop=my_font)
plt.show()
```

```python
cor = 1.0
i = 0
shape = ['-k', '--k', '-.k', ':k']
for loss in [train_loss_8_0,train_loss_16_0,train_loss_32_0,train_loss_64_0]:
    plt.plot([x for x in range(100)],[float(x) for x in loss], shape[i] ,label='λ=%s'%str(cor))
    i = i + 1
    cor = cor * 2

plt.xlabel('训练轮次',FontProperties=my_font)
plt.ylabel('训练集损失',FontProperties=my_font)
plt.legend(prop=my_font)
plt.show()
```

```python
cor = 1.0
i = 0
shape = ['-k', '--k', '-.k', ':k']
for loss in [test_accuracy_8_0,test_accuracy_16_0,test_accuracy_32_0,test_accuracy_64_0]:
    plt.plot([x for x in range(100)],[float(x) for x in loss], shape[i] ,label='λ=%s'%str(cor))
    i = i + 1
    cor = cor * 2

plt.xlabel('训练轮次',FontProperties=my_font)
plt.ylabel('测试集准确率',FontProperties=my_font)
plt.legend(prop=my_font)
plt.show()
```
